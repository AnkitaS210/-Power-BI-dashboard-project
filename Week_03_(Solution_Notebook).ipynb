# Data Cleaning - 1

# Column Names (upper to lower case)
*   columns = to access all the columns
*   str = to convert the data to a string
*   lower = convert to lower case

```python
# changing col names from uppercase to lowercase:
data.columns.str.lower()

# as we want to replace ' ' (space) with '_' (underscore) then:
data.columns = data.columns.str.lower().str.replace(' ', '_')

# also assigned the changes to the original data with variable name 'data.columns'
```

```python
# printing updated dataset
data.head()
```

# Index Column: dropping this column

*   axis = 1 : indicating that the operation should be applied to columns
*   axis = 0 : indicating that the operation should be applied to rows

*   inplace = True : modifies the existing data frame and does not creates a new one

```python
# dropping the column named 'index'

data = data.drop('index', axis=1)
```

```python
# checking the updations

data.head()
```

# Age Column: Treat missing values

```python
# accessing age column (can be done using two methods)

# first method:
data['age']

# second method:
data.age
```

```python
# calculating mean: finding avg age of a person in age column

# MEAN:
avg_age = data['age'].mean()
avg_age

# we can find Meadin and Mode (most frequent value) too: [but here for treating missing values we'll go with mean value]

# MEDIAN:
# data['age'].median()

# MODE:
# data['age'].mode()
```

```python
# fill the missing values with the calculated mean (there are 7 missing values in this col)

data['age'] = data.age.fillna(avg_age)
# fill.na is a method to fill missing values

# data['age'] = : making changes to this col in the original data is done by assigning it with a variable
# do not replace the updates to the whole data replace it to a particular col (like for e.g. 'Age' col)

# lets round up the decimals

# to remove all the decimals:
data['age'] = data.age.round().astype(int)
# astype(int) is used to convert rounded values to integer

# to keep one decimal:
# data['age'] = data.age.round(decimals = 1)
```

```python
# checking the updations

data.head()
```

# Salary Column

```python
data['salary'] = data.salary.str.replace('$', '', regex = True).str.replace('k', '000', regex = True)
```

```python
# checking the updations

data.head()
```

# Rating column

```python
# finding median for treating missing values

r_median = data.rating.median()
r_median
```

```python
data['rating'] = data.rating.fillna(r_median).replace(-1.0, r_median)
```

```python
# checking the updations

data.head()
```

# Location column

```python
data.location.value_counts()
```

```python
data['location'] = data.location.str.replace(',', '').str.rstrip('InNy Aus')
```

```python
# checking the updations

data.head()
```

# Established column

```python
data['established'] = data.established.replace(-1, 'Unknown')
```

```python
# checking the updations

data.head()
```

# Easy_apply column

```python
data['easy_apply'] = data.easy_apply.str.replace('-1', 'FALSE').str.title()
```

```python
# checking the updations

data.head()
```
